```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function for the convex shape (bowl)
def convex_function(x, y):
    return x**2 + y**2

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute the z values (cost function)
z = convex_function(x, y)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)

# Add labels and title
ax.set_xlabel('Theta 0 (θ₀)')
ax.set_ylabel('Theta 1 (θ₁)')
ax.set_zlabel('Cost (J)')
ax.set_title('Convex Cost Function (Bowl Shape)')

# Add a color bar for reference
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```
